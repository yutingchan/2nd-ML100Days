```python
import os #為處理作業系統相關及目錄路徑的模組
import numpy as np #Python進行科學計算的一個基礎包
import pandas as pd 
"""
pandas是資料分析包
pandas中主要有兩種資料結構，分別是：Series和DataFrame
Series：一種類似於一維陣列的物件，是由一組資料(各種NumPy資料型別)以及一組與之相關的資料標籤 (即索引)組成。
DataFrame：一個表格型的資料結構，包含有一組有序的列，每列可以是不同的值型別 (數值、字串、布林型等)。
"""
```

```python
##os常用函數(function)
#取得目前所在路徑
os.getcwd()
#列出當前路徑中所有資料
os.listdir(path='.')
#建立路徑
os.mkdir(path, mode=0o777, *, dir_fd=None)
#刪除目錄
os.rmdir(path, *, dir_fd=None)
#判斷path檔案是否存在
os.path.isfile(path)
#判斷path路徑是否存在
os.path.isdir(path)
#分割路徑為目錄(head)及檔案名稱(tail)
os.path.split(path)
#分割路徑為路徑加檔名(root)及副檔名(ext)
os.path.splitext(path)
#回傳 path 的目錄路徑
os.path.dirname(path)
#回傳 path 的檔案名稱
os.path.basename(path)
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
#將 dirpath 與 filename 結合，無需自行判斷要用 '\\' 或 '/'
f_app = os.path.join(dir_data, 'application_train.csv')#可解決不同系統之符號不同問題，如windows系統路徑分隔符號為\，mac為/

print('Path of read in data: %s' % (f_app)) # %s 以str()輸出
app_train = pd.read_csv(f_app)
```

#### 截取部分資料

```python
#檢視第1~20筆資料
app_train[1:20]
#檢視某個變項資料
app_train["CODE_GENDER"]
#篩選性別為女性的資料
filter = app_train["CODE_GENDER"] == 'F'
app_train[filter]
#多重條件篩選
app_train[(app_train["CODE_GENDER"] == 'F') & (app_train["AMT_INCOME_TOTAL"] > 135000)] # |:or &:and
#找出NA值的位置
app_train["AMT_REQ_CREDIT_BUREAU_HOUR"].isnull()
#刪除NA值
app_train.dropna()
#篩選出女性並指顯示其中感興趣幾個欄位資料
app_train[['CODE_GENDER','AMT_INCOME_TOTAL']][app_train["AMT_INCOME_TOTAL"] > 135000]
```
